<h3>Task 1</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Generate noisy data
X, y = make_moons(n_samples=150, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Plot the raw data
plt.figure(figsize=(8, 5))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='RdBu', edgecolors='k', label='Train Data')
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, cmap='RdBu', marker='x', alpha=0.5, label='Test Data')
plt.title("The Raw Data: Non-Linear and Noisy")
plt.legend()
plt.show()
```

```python
# preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Transformation polynomiale degré 10
poly = PolynomialFeatures(degree=10)
X_poly = poly.fit_transform(X_scaled)
```

```python
# 4. Modèle de classification sans reguliarisation
model = LogisticRegression(penalty=None, max_iter=10000)
model.fit(X_poly, y)
```

```python
# Visualisation de la frontière de décision
x1 = np.linspace(-2, 2, 200)
x2 = np.linspace(-2, 2, 200)
xx, yy = np.meshgrid(x1, x2)

X_new = np.c_[xx.ravel(), yy.ravel()]
X_new_scaled = scaler.transform(X_new)
X_new_poly = poly.transform(X_new_scaled)

Z = model.predict(X_new_poly)
Z = Z.reshape(xx.shape)
```

```python
# Plot
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.title("Frontière de décision (degré 10)")
plt.show()
```

```python
# Régularisation Ridge (L2)
model_ridge = LogisticRegression(penalty='l2', C=1.0, max_iter=10000)
model_ridge.fit(X_poly, y)
```

```python
# Régularisation Lasso (L1)
model_lasso = LogisticRegression(penalty='l1', solver='liblinear', C=1.0, max_iter=10000)
model_lasso.fit(X_poly, y)
```

```python
models = {
    "Sans régularisation": LogisticRegression(penalty=None, max_iter=10000),
    "Ridge (L2)": LogisticRegression(penalty='l2', C=1.0, max_iter=10000),
    "Lasso (L1)": LogisticRegression(penalty='l1', solver='liblinear', C=1.0, max_iter=10000)
}

plt.figure(figsize=(15, 4))

for i, (title, model) in enumerate(models.items()):
    model.fit(X_poly, y)

    Z = model.predict(X_new_poly)
    Z = Z.reshape(xx.shape)

    plt.subplot(1, 3, i+1)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title(title)

plt.tight_layout()
plt.show()
```

<h3>Task 2</h3>

```python
from sklearn.datasets import make_classification
```

```python
# 1. Génération du dataset (3k samples, 50 features)
X, y = make_classification(n_samples=3000, n_features=50, n_informative=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 2. Prétraitement
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 3. Comparaison des degrés et des modèles
degrees = [2, 3]
results = {}

for deg in degrees:
    print(f"Calcul pour le degré {deg}...")
    poly = PolynomialFeatures(degree=deg)
    X_train_poly = poly.fit_transform(X_train_scaled)
    X_test_poly = poly.transform(X_test_scaled)
    
    models = {
        "Sans régul": LogisticRegression(penalty=None, max_iter=100),
        "Ridge (L2)": LogisticRegression(penalty='l2', C=0.1, max_iter=100),
        "Lasso (L1)": LogisticRegression(penalty='l1', solver='liblinear', C=0.1, max_iter=100)
    }
    
    for name, model in models.items():
        model.fit(X_train_poly, y_train)
        acc = accuracy_score(y_test, model.predict(X_test_poly))
        results[f"Deg {deg} - {name}"] = acc
```

```python
# 4. Visualisation de l'Accuracy par rapport aux Itérations (Epochs)
# On prend le cas le plus complexe : Degré 2 avec Ridge
iters = range(1, 21, 2)
train_acc = []
test_acc = []

# On utilise un degré 2 pour que l'exemple soit fluide
poly_v2 = PolynomialFeatures(degree=2)
X_tr_p = poly_v2.fit_transform(X_train_scaled)
X_te_p = poly_v2.transform(X_test_scaled)

model_iter = LogisticRegression(penalty='l2', solver='saga', warm_start=True, max_iter=1)

for i in iters:
    model_iter.max_iter = i
    model_iter.fit(X_tr_p, y_train)
    train_acc.append(accuracy_score(y_train, model_iter.predict(X_tr_p)))
    test_acc.append(accuracy_score(y_test, model_iter.predict(X_te_p)))
```

```python
# --- AFFICHAGE DES RÉSULTATS ---

# Graphique de l'évolution de l'apprentissage
plt.figure(figsize=(10, 5))
plt.plot(iters, train_acc, label="Train Accuracy", marker='o')
plt.plot(iters, test_acc, label="Test Accuracy", marker='s')
plt.title("Évolution de l'Accuracy par Itérations (Degré 2 + Ridge)")
plt.xlabel("Itérations (équivalent époques)")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

Le graphique montre une divergence entre l'accuracy d'entraînement et de test. Alors que l'accuracy d'entraînement atteint la perfection (1.0), l'accuracy de test plafonne puis régresse. Cela démontre que le modèle de degré 2 entre en phase de sur-apprentissage (overfitting) dès la 5ème itération. La régularisation Ridge utilisée ici limite l'explosion des poids mais ne suffit pas totalement à compenser la haute dimensionnalité du modèle.

```python
# Comparaison finale
print("\n--- SCORE FINAL SUR TEST SET ---")
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```
